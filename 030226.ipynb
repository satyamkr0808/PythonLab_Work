```python
def dfs(st, goal, graph, vis = None):
    if vis is None:
        vis = set()
    print(f"Visiting: {st}")
    if st == goal:
        return True
    vis.add(st)
    for child in graph.get(st, []):
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True
    return False

graph = {
         'A': ['B', 'C'],
         'B': ['A', 'C', 'D'],
         'C': ['A', 'B', 'E'],
         'D': ['B', 'E', 'F'],
         'E': ['C', 'D', 'F'],
         'F': ['D', 'E']
        }

result = dfs('A', 'F', graph)
print(f"Path found: {result}")
```

```python
def dfs(s, goal, graph):
    res=[]
    st=[]
    vis=set()
    st.append(s)
    while(len(st)!=0):
        u=st.pop()
        if(u not in vis):
            vis.add(u)
            res.append(u)
            if(u==goal):
                return res
            for v in graph[u]:
                st.append(v)
    return res;

graph= {'A': ['B', 'C'],
 'B': ['A', 'C', 'D'],
 'C': ['A', 'B', 'E'],
 'D': ['B', 'E', 'F'],
 'E': ['C', 'D', 'F'],
 'F': ['D', 'E']}
result = dfs('A', 'F', graph)
print(f"Path found: {result}") 
```
