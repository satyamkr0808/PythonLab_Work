```python
def dfs(st, goal, graph, vis = None):
    if vis is None:
        vis = set()
    print(f"Visiting: {st}")
    if st == goal:
        return True
    vis.add(st)
    for child in graph.get(st, []):
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True
    return False

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'D'],
    'D': ['B', 'C', 'E'],
    'E': ['B', 'D']
}

result = dfs('A', 'E', graph)
print(f"Path found: {result}")
```

```python
def dfs(s, goal, graph):
    res=[]
    st=[]
    vis=set()
    st.append(s)
    while(len(st)!=0):
        node=st.pop()
        if(node not in vis):
            vis.add(node)
            res.append(node)
            if(node==goal):
                return res
            for neighbour in graph[node]:
                st.append(neighbour)
    return res

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'D'],
    'D': ['B', 'C', 'E'],
    'E': ['B', 'D']
}

result = dfs('A', 'E', graph)
print(f"Path found: {result}")
```
