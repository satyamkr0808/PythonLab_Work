```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
rooms = np.random.randint(1, 6, 20)
length = np.random.randint(20, 60, 20)
breadth = np.random.randint(15, 50, 20)
area = length * breadth
Price = 50000 + (area * 40) + (rooms * 250) + np.random.randint(0, 150, 20)

df = pd.DataFrame({
    'room': rooms,
    'length': length,
    'breadth': breadth,
    'area': area,
    'Price': Price
})
print(df.head(5))
```

```python
X = df[['area','length','breadth','room']]
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
print(f'''\t\tX_train:\n{X_train},\n
\t\tX_test:\n{ X_test},\n
\t\tY_train:\n{ y_train},\n 
\t\tY_test:\n{ y_test}
''')
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
print()
print(y_test)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope:\n{slope}\nIntercept:\n{intercept}")
```

```python
print(f'''MAE:{mean_absolute_error(y_test, y_pred)}
MSE:{mean_squared_error(y_test, y_pred)}
RMSE:{np.sqrt(mean_squared_error(y_test, y_pred))}
R2_SCORE:{r2_score(y_test, y_pred)}
''')
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
num_rooms = np.random.randint(1, 7,20)
categories = ['big apartment' if r >= 3 else 'small' for r in num_rooms]

df = pd.DataFrame({
    'Rooms': num_rooms,
    'category': categories
})
print(df.head())
```

```python
X = df[['Rooms']].head()
y = df[['category']].head()
```

```python
print(f"{X},\n {y}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

#MODEL TRAIN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred=knn.predict(X_test)
print(y_pred)
print()
print(y_test)
```

```python
print(f''' Accuracy: {accuracy_score(y_test, y_pred)}
 Precision: {precision_score(y_test, y_pred,average='weighted')}
 Recall: {recall_score(y_test, y_pred,average='weighted')}
 F1-Score: {f1_score(y_test, y_pred,average='weighted')}''')
```

```python
#CONFUSION MATRIX
print(confusion_matrix(y_test, y_pred))
```

```python
#CLASSIFICATION REPORT
print(classification_report(y_test, y_pred))
```
